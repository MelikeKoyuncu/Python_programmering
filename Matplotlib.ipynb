### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

In Matplotlib, there are two ways to create plots; the OO-approach and the pyplot/MATLAB approach. The OO-approach is recommended for plots which are more complicated and part of a bigger project. It provides more control and customization of the plots. And pyplot/MATLAB approach is a more state based way of creating plots. This approach is recommended for quick interactive work.

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

Figure is the top-level container for all plot elements and can be seen as canvas. 
Axes is the part of the “canvas” that the plot is attached to is called axes. It is a container that holds the plot elements such as lines,markers,text, etc.
Axis object contains the axis label, tick labels, tick locations, and grid lines.
Artist is any object that can be drawn using the renderer.

### When plotting in Matplotlib, what is the expected input data type?

When plotting in Matplotlib, the expected input data type is a numpy.array. Classes that are similar to arrays ('array-like') such as pandas data objects.

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
# OO-approach:

# Define the range of x values
x = np.linspace(-4, 4, 100)

# Calculate the corresponding y values
y = x ** 2

# Create a figure and axis object
fig, ax = plt.subplots()

# Plot the function
ax.plot(x, y)

# Set the title and axis labels
ax.set_title('y = x^2')
ax.set_xlabel('x')
ax.set_ylabel('y')
```

```python
# Pyplot approach:

# Plot the function
plt.plot(x, y)

# Set the title and axis labels
plt.title('y = x^2')
plt.xlabel('x')
plt.ylabel('y')
```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Data for bar plot
fruit_data = {'Grapes': 22, 'Apple': 8, 'Orange': 15, 'Lemon': 20, 'Lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())

# Create figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2,layout='constrained')
fig.suptitle('Different Plots', size = 20)

# Plot scatter plot on first subplot
ax1.scatter(random_data_x, random_data_y)

# Plot bar plot on second subplot
ax2.bar(names, values)

# Set titles and labels for subplots and figure
ax1.set_title('Scatter Plot')
ax2.set_title('Bar Plot')
```
